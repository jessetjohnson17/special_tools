```python
import pandas as pd
import folium
import matplotlib.pyplot as plt

df = pd.read_csv('incident_records.csv')
```

```python
df.head()
```

```python
center_lat = df["latitude"].mean()
center_lon = df["longitude"].mean()
m = folium.Map(location=[center_lat, center_lon], zoom_start=2)

# Add one marker per row, with a popup and tooltip
for _, row in df.iterrows():
    folium.Marker(
        location=[row["latitude"], row["longitude"]],
        popup=row["country_txt"],
        tooltip="Click for country name"
    ).add_to(m)

#m
```

```python
from folium.plugins import HeatMap

hm = folium.Map(location=[center_lat, center_lon], zoom_start=2)
HeatMap(df[['latitude', 'longitude']], max_opacity = .8, blur=15, radius=10).add_to(hm)
hm

```
